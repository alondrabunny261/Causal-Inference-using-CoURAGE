```python
from itertools import combinations

import numpy as np
from scipy import stats

import networkx as nx
import pandas as pd

import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
COLORS = [
    '#00B0F0',
    '#FF0000'
]
```

```python
data = pd.read_csv("M1_meteorological_data.csv")

precipitation = data['precipitation']

# `a` and `c` are causally independent of each other, but they are both children of `b` 
specific_humidity = precipitation + data['specific_humidity']
temperature = precipitation + data['temperature']
```

```python
# Let's check correlation between `a` and `c`
coef, p_val = stats.pearsonr(specific_humidity, temperature)

print(coef)
```

```python
variables = {
    's-humidity': specific_humidity,
    'precipitation': precipitation,
    'temperature': temperature
}

plt.figure(figsize=(12, 7))

for i, (var_1, var_2) in enumerate([('precipitation', 's-humidity'), 
                                    ('precipitation', 'temperature'), 
                                    ('temperature', 's-humidity')]):
    
    color = COLORS[0]
    
#     if 'precipitation' in [var_1, var_2]:
#         color = COLORS[0]
    
    plt.subplot(2, 2, i + 1)
    plt.scatter(variables[var_1], variables[var_2], alpha=.8, color=color)

    plt.xlabel(f'${var_1}$', fontsize=16)
    plt.ylabel(f'${var_2}$', fontsize=16)

plt.suptitle('Pairwise relationships between $Specific$ $Humidity$, $Precipitation$ and $Temperature$')
plt.subplots_adjust(hspace=.25, wspace=.25)
plt.show()
```
