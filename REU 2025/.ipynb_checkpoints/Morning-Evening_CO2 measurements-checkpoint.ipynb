```python
# Hours of interest(Peak) 5->10 , 3->8

data = pd.read_csv(f"{site}_30_min_meteorological_CO2_aerosol_data.csv")
```

```python
data['time'] = pd.to_datetime(data['time'])

df_peak_morning = data[
    ((data['time'].dt.hour == 5)) |
    ((data['time'].dt.hour == 6)) |
    ((data['time'].dt.hour == 7)) |
    ((data['time'].dt.hour == 8)) |
    ((data['time'].dt.hour == 9)) |
    ((data['time'].dt.hour == 10) & (data['time'].dt.minute == 0))
]

df_peak_evening = data[
    ((data['time'].dt.hour == 15)) |
    ((data['time'].dt.hour == 16)) |
    ((data['time'].dt.hour == 17)) |
    ((data['time'].dt.hour == 18)) |
    ((data['time'].dt.hour == 19)) |
    ((data['time'].dt.hour == 20) & (data['time'].dt.minute == 0))]
```

```python
for i in df_peak_morning:
    print(i)
```

```python
df_peak_morning['co2 flux']
```

```python
# Step 2: Extract the hour
df_peak_morning['hour'] = df_peak_morning['time'].dt.hour + df_peak_morning['time'].dt.minute / 60  # support for half-hour

# Step 3: Plot
plt.figure(figsize=(14, 8))
scatter = plt.scatter(
    df_peak_morning['time'],
    df_peak_morning['co2 spec. humidity (meas)'],
    c=df_peak_morning['hour'],
    cmap='turbo',  
    s=5
)

# Step 4: Add colorbar
cbar = plt.colorbar(scatter, label='Hour of Day')
#plt.ylim(10,30)
plt.xlabel("Time")
plt.ylabel("CO₂ Flux")
plt.title("CO₂ Flux Colored by Hour of Day")
plt.tight_layout()
plt.show()
```

```python
# Step 2: Extract the hour
df_peak_evening['hour'] = df_peak_evening['time'].dt.hour + df_peak_evening['time'].dt.minute / 60  # support for half-hour

# Step 3: Plot
plt.figure(figsize=(14, 8))
scatter = plt.scatter(
    df_peak_evening['time'],
    df_peak_evening['co2 spec. humidity (meas)'],
    c=df_peak_evening['hour'],
    cmap='magma',  
    s=5
)

# Step 4: Add colorbar
cbar = plt.colorbar(scatter, label='Hour of Day')
#plt.ylim(-50,50)
plt.xlabel("Time")
plt.ylabel("CO₂ Flux")
plt.title("CO₂ Flux Colored by Hour of Day")
plt.tight_layout()
plt.show()
```
