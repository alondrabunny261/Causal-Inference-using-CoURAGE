```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

df = pd.read_csv("M1_30_min_meteorological_data.csv")

df = df.drop('relative humidity', axis=1)
df = df.drop('dew_point', axis=1)
df = df.drop('wind speed', axis=1)
df = df.drop('atmospheric pressure', axis=1)
df = df.drop('wetness', axis=1)
df = df.drop('net radiation', axis=1)
df = df.drop('albedo', axis=1)
#data = data.drop('precipitation', axis=1)

# Asegúrate de que la columna 'time' sea tipo datetime
df['time'] = pd.to_datetime(df['time'])

# Lista de variables (excluyendo 'time')
variables = [col for col in df.columns if col != 'time']
met_vars = ['Precipitation','Temperature (°C)', "Specific Humidity (g/kg)"]

# Crear subplots
fig, axes = plt.subplots(len(variables), 1, figsize=(12, 8), sharex=True)

# Asegurar que axes sea iterable (por si solo hay una variable)
if len(variables) == 1:
    axes = [axes]

# Colores (reutilizables)
colors = ['orchid', 'skyblue', 'pink', 'orange', 'purple', 'brown']

# Graficar cada variable
for i, var in enumerate(variables):
    axes[i].plot(df['time'], df[var], color=colors[i % len(colors)])
    axes[i].set_ylabel(met_vars[i], fontsize=14)
    #axes[i].grid(True, linestyle='--', alpha=0.4)

# Formato de fechas en eje X
axes[-1].xaxis.set_major_formatter(mdates.DateFormatter('%b-%Y'))  # Ej: Jan-2020
axes[-1].xaxis.set_major_locator(mdates.MonthLocator(interval=1))  # Cada mes

# Rotar etiquetas
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
fig.align_ylabels(axes)
plt.show()
```

```python
#From metpy package: Create a new column for data that contains dew point and specific humidity
# Calculate dew point using your formula : T - ((100- RH)/5)
data_balt['dew_point'] = data_balt['temp_mean'] - ((100 - data_balt['rh_mean']) / 5)

# Convert pressure from kPa to hPa
pressure = (data_balt['atmos_pressure'] * 10).values * units.hPa

# Convert dew point to proper units
dew_point = data_balt['dew_point'].values * units.degC

# Compute specific humidity
sh = shfd(pressure, dew_point).to('g/kg')

# Add to dataframe
data_balt['specific_humidity'] = sh.magnitude  # only store the values, not units

# Save updated CSV
data_balt.to_csv('temp_rh_precip_combined.csv', index=False)
```

```python
# Ensure datetime format
data_balt['time'] = pd.to_datetime(data_balt['time'])

# Bin time to 30-minute intervals
data_balt['time_bin'] = data_balt['time'].dt.floor('30min')

# Map each bin to a code for color
data_balt['time_bin_code'] = data_balt['time_bin'].astype('category').cat.codes

# Fit linear regression line
slope, intercept = np.polyfit(data_balt.temp_mean, data_balt.specific_humidity, 1)

# Plot
plt.figure(figsize=(10, 10))
sc = plt.scatter(
    data_balt.temp_mean,
    data_balt.specific_humidity,
    c=data_balt.time_bin_code,
    cmap='gist_ncar',
    alpha=0.8
)

plt.plot(
    data_balt.temp_mean,
    data_balt.temp_mean * slope + intercept,
    color="blue",
    label=f"y = {slope:.2f}x + {intercept:.2f}"
)

plt.xlabel("Temperature (°C)")
plt.ylabel("Specific humidity (g/kg)")
plt.title(f"Specific humidity vs Temperature: Day {data_balt.base_time[0]}")
plt.text(16, 7, f"y = {slope:.2f}x + {intercept:.2f}", fontsize=12, color='blue')

# Colorbar with actual times
cbar = plt.colorbar(sc)
cbar.set_label("Time (30-min bins)")

# Optional: Format colorbar ticks to show human-readable times
unique_times = data_balt['time_bin'].dt.strftime('%H:%M').unique()
tick_locs = np.linspace(0, len(unique_times) - 1, num=6, dtype=int)
cbar.set_ticks(tick_locs)
cbar.set_ticklabels(unique_times[tick_locs])

plt.savefig("Specific_humidity_by_half_hour.png")
plt.show()
```

```python
#CSV files
csv_files = sorted([f for f in os.listdir() if f.endswith(".csv")])

all_times = []
all_temps = []
all_rh = []
all_wds = []
days = []
temp_mean_daily = []
rh_mean_daily = []
wds_mean_daily = []

minutes_per_day = 1440

for i, file in enumerate(csv_files):
    data = pd.read_csv(file)
    all_times.extend(data.time)
    all_temps.extend(data.temp_mean)
    all_rh.extend(data.rh_mean)
    all_wds.extend(data.wspd_vec_mean)
    days.append(data.base_time[0])
    
    temp_mean_daily.append(np.format_float_positional(np.mean(data.temp_mean), precision=3))
    rh_mean_daily.append(np.format_float_positional(np.mean(data.rh_mean), precision=3))
    wds_mean_daily.append(np.format_float_positional(np.mean(data.wspd_vec_mean), precision=3))

# Numpy arrays
all_times = np.array(all_times)
all_temps = np.array(all_temps)
all_rh = np.array(all_rh)
all_wds = np.array(all_wds)

all_times_dt = pd.to_datetime(all_times)

# Atmospheric variables
variables = {
    "Temperature (°C)": (all_temps, temp_mean_daily, "magenta"),
    "Relative Humidity": (all_rh, rh_mean_daily, "royalblue"),
    "Wind Speed (m/s)": (all_wds, wds_mean_daily, "green"),
}

def plot_variable(var_name):
    values, mean_daily, color = variables[var_name]

    # Filter for 2025
    mask_2025 = all_times_dt.year == 2025
    filtered_times = all_times_dt[mask_2025]
    filtered_values = np.array(values)[mask_2025]
    
    days_dt = pd.to_datetime(days)
    day_centers = [all_times_dt[0] + pd.Timedelta(minutes=minutes_per_day * i + minutes_per_day // 2) for i in range(len(days))]
        
    hover_texts = [f"{var_name}: {v:.2f}<br>Time: {t} <br> Mean: {mean_daily[days_dt.get_indexer([t.floor('D')])[0]]}" for v, t in zip(filtered_values, filtered_times)]

    fig = go.Figure()
    
    # Add vertical lines at day centers
    for center in day_centers:
        fig.add_vline(
            x=center,
            line=dict(color='gray', dash='dash'),
            layer='below')

    fig.add_trace(go.Scatter(
        x=filtered_times,
        y=filtered_values,
        mode='lines+markers',
        line=dict(color=color),
        marker=dict(size=3),
        text=hover_texts,
        hoverinfo='text',
        name=var_name
    ))

    fig.update_layout(
        title=f"{var_name}",
        xaxis_title="Time",
        yaxis_title=var_name,
        hovermode='closest',
        height=500
    )
    fig.show()

    
# Widget 
interact(plot_variable, var_name=widgets.Dropdown(options=list(variables.keys()), value="Temperature (°C)", description="Variable:"))
```
