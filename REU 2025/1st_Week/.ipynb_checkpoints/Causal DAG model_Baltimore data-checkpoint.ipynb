# Real data aproximations

```python
df = pd.read_csv('crgmetM1.b1.20250528.000000.custom.csv')
df = df[['org_precip_rate_mean', 'rh_mean', 'temp_mean', 'base_time','time']]
```

```python
# Identify the causal effect

# Digraph Syntax
causal_graph = """digraph {org_precip_rate_mean -> rh_mean; 
                           org_precip_rate_mean -> temp_mean; 
                           rh_mean -> temp_mean;}"""

model= CausalModel(data=df, # actual dataset now
        graph=causal_graph, 
        treatment='org_precip_rate_mean', #causal, x
        outcome=['rh_mean', 'temp_mean']) # cause of interest,  # outcome, Y

model.view_model()
```

# Estimation

```python
estimands = model.identify_effect()
print(estimands)
```

```python
estimate = model.estimate_effect(estimands,
                                 method_name="backdoor.linear_regression",
                                 effect_modifiers= [],
                                 confidence_intervals=True,
                                 test_significance=True)

print(estimate)
```

```python
estimate.interpret()
#Great, this results in what we thought due to intuition: More rh, less t 
```

```python

plt.figure(figsize=(10,10))
sns.regplot(x='rh_mean',y='org_precip_rate_mean', data=df, scatter_kws={'color': 'skyblue'}, line_kws={'color': 'red'})
plt.xlabel("Mean Temperature")
plt.ylabel("Relative Humidity")
plt.title(f"Linear Relationship: Relative Humidity vs Precipitation - Day: {df.base_time[0]}")
plt.show()
```

```python
#Regression standard python model
#Effects on rh due to precip: rh = m(precip) + b
reg_model_all = smf.ols(formula='rh_mean ~ org_precip_rate_mean', data = met_df)

ols_all = reg_model_all.fit()
print('Effect of Precipitation on RH: \n coeff: \t', ols_all.params['org_precip_rate_mean'], 
      '\n p value: \t', ols_all.pvalues.loc['org_precip_rate_mean'], 
     '\n 95% C.I.s: \t', ols_all.conf_int(alpha = 0.05).loc['org_precip_rate_mean'].values)

print(ols_all.summary())
```

```python
# Results from ols_all.summary()
coefs = ols_all.params

#[0.025 -> 0.975]
conf = ols_all.conf_int()

#Standar Deviation
errors = ols_all.bse


# Variables
x = met_df['org_precip_rate_mean']  # Precipitación
y = met_df['rh_mean']               # Humedad relativa
time= met_df['time']

x_mean=np.mean(x)
print(x_mean)
y_mean=np.mean(y)
print(y_mean)

# Coeficients for Linear regression line
intercept = coefs[0]
slope = coefs[1]

# Línea de regresión
x_line = np.linspace(x.min(), x.max(), 100)
y_line = slope * x_line + intercept

######## Plot #################
plt.figure(figsize=(8, 6))
plt.scatter(x, y, alpha=0.6, color='cornflowerblue', label='Precipitation')
plt.plot(x_line, y_line, color='darkred', linewidth=2, label='Linear Regression')
plt.plot(x_mean,y_mean, color="magenta", marker="*",ms=10)
plt.text(1,70, f"y={slope}x+{intercept}", bbox=dict(boxstyle="round,pad=0.2", fc="white", ec="gray"))
plt.ylim(25,102)
plt.xlim(0,20)

plt.xlabel('Precipitation')
plt.ylabel('Relative Humidity')
plt.title('Relation: Precipitation and Relative Humidity Linear Regression')
plt.legend()
#plt.grid(True)
plt.tight_layout()
plt.show()
```
