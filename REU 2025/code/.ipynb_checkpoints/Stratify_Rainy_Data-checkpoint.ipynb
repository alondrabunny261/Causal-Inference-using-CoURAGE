```python
site = "M1"
path = f"{site}_data/1_min_aerosols/"

#data =pd.read_csv(path + f'{site}_minute_meteorological_data_wxt.csv')

data = pd.read_csv(path + f"{site}_minute_meteorological_with_aerosols_data_wxt.csv")

data.index = pd.to_datetime(data['time'])

# Meteorological variables
#precipitation = data['wxt_precip_rate_mean']
precipitation = data['precipitation']
specific_humidity = data['specific humidity']
```

```python
rain = np.ma.masked_where(precipitation <= 0, precipitation)

no_rain = np.ma.masked_where(precipitation > 0.0000, precipitation)
```

```python
########################### Rain ###########################

m1_day = []
m1_precipitation = []
m1_relative_humid = []
m1_temperature = []
m1_atmos_pressure = []
m1_winds = []
m1_specific_humid = []
m1_dew_point = []

for a, i in enumerate(rain):
    if not rain.mask[a]:
        m1_day.append(data['time'][a])
        m1_precipitation.append(data['precipitation'][a])
        #m1_precipitation.append(data['wxt_precip_rate_mean'][a])
        m1_relative_humid.append(data['relative humidity'][a])
        m1_temperature.append(data['temperature'][a])
        m1_atmos_pressure.append(data['atmos_pressure'][a])
        m1_winds.append(data['wspd_arith_mean'][a])
        m1_specific_humid.append(data['specific humidity'][a])
        m1_dew_point.append(data['dew_point'][a])

    
df = pd.DataFrame({'time': m1_day,
                   'precipitation': m1_precipitation,
                   'relative humidity': m1_relative_humid,
                   'temperature': m1_temperature,
                   'atmospheric pressure': m1_atmos_pressure,
                   'wind speed': m1_winds,
                   'specific humidity': m1_specific_humid,
                   'dew point' : m1_dew_point,
                  })
        
    
df.to_csv(path + f"{site}_30_min_rain_minute_meteorological_data.csv", index=False)
  
```

```python
# Time from dataframe
data_rain['time'] = pd.to_datetime(data_rain['time'])
data_rain['date'] = data_rain['time'].dt.date

# Amount of events per day
counts_per_day = data_rain.groupby('date')['time'].count()

full_days = counts_per_day[counts_per_day >= 20].index
continuous_rain = data_rain[data_rain['date'].isin(full_days)].copy()

# Sorts time in a query
continuous_rain = continuous_rain.sort_values('time')

# Small Interruption
max_gap = pd.Timedelta(minutes=10)

# Gap from small interruption
continuous_rain['gap'] = continuous_rain['time'].diff() > max_gap

continuous_rain['group'] = continuous_rain['gap'].cumsum()

grouped = continuous_rain.groupby('group')
valid_groups = grouped.filter(lambda x: len(x) > 1)

rain_periods = valid_groups.groupby('group')['time'].agg(['min', 'max'])
rain_periods.columns = ['start_time', 'end_time']

# Duration
rain_periods['duration_min'] = (
    (rain_periods['end_time'] - rain_periods['start_time']).dt.total_seconds() / 60
)

# Only events higher or equal than 20 mins
rain_periods = rain_periods[rain_periods['duration_min'] >= 20]

# Save
#rain_periods.to_csv(path + f"{site}_continuous_rain_periods.csv", index=False)
```

```python
############################# Plotting Events ########################### 
fig, ax = plt.subplots(figsize=(10, 4))

plt.scatter(data_rain.time ,data_rain.precipitation,color="teal", s=5)

for i, row in rain_periods.iterrows():
    ax.axvspan(row['start_time'], row['end_time'], color='lightcoral', alpha=0.8)

ax.set_xlabel("Time", fontsize=15)
ax.set_ylabel("Precipitation rate (mm/h)", fontsize=15)
ax.set_title("Rural Facility: Precipitation Periods (≥ 20 min)", fontsize=18)
plt.xticks(rotation=45, fontsize=15)
plt.tight_layout()
plt.show()
```
