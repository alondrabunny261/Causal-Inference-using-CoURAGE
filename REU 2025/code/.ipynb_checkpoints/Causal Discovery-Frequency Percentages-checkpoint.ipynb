```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from ast import literal_eval

# Tus columnas
columns = ['P_P_elem', 
           'P_T_elem', 
           'P_H_elem',
           'T_P_elem',
           'T_T_elem',
           'T_H_elem',
           'H_P_elem',
           'H_T_elem',
           'H_H_elem']

# Total de muestras
n_samples = len(data_with_lags)

# Inicializar grafo dirigido
G = nx.DiGraph()

# Diccionario para contar ocurrencias significativas
relation_counts = {}

for col in columns:
    count = 0
    for lag_str in data_with_lags[col]:
        lags = literal_eval(lag_str)
        if len(lags) > 0:
            count += 1
    relation_counts[col] = count

# Crear el grafo solo con relaciones significativas >= 50%
for col, count in relation_counts.items():
    if count / n_samples >= 0.3:
        source, target = col.replace('_elem', '').split('_')
        # Aadir flecha con etiqueta (puedes usar el lag promedio tambin si quieres)
        G.add_edge(source, target, label=f"{count}/{n_samples}")

# Dibujar
pos = nx.spring_layout(G, seed=42)
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_size=2000, node_color="lightblue", arrows=True, font_weight='bold')

edge_labels = nx.get_edge_attributes(G, 'label')

nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.title(f"{site}: DAG con relaciones significativas  30%")
plt.show()
```

```python
n_samples = len(data_with_lags)

results = []

# Iterar por cada relacin
for col in columns:
    all_lags = []

    for lag_str in data_with_lags[col]:
        lags = literal_eval(lag_str)
        all_lags.extend(lags)

    lag_counter = Counter(all_lags)

    # Guardar solo los lags que aparecen en  1% de las muestras
    for lag_value, count in lag_counter.items():
        percentage = (count / n_samples) * 100
        if percentage >= 1:
            relation = col.replace('_elem', '').replace('_', '  ')
            results.append({
                'Relation': relation,
                'Lag': lag_value,
                'Percentage': round(percentage, 2),
                'Count': count,
                'Total Samples': n_samples
            })

# Convert to DataFrame
df_results = pd.DataFrame(results)

# Save to CSV
#df_results.to_csv(path + f'{site}_lag_summary_morning.csv', index=False)
df_results.to_csv(path + f'{site}_lag_summary_rain.csv', index=False)
```
