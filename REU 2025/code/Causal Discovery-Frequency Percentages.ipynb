```python

# columns = ['P_P_elem', 
#            'P_T_elem', 
#            'P_H_elem',
#            'P_CO2_elem',
#            'P_aero_elem',
           
#            'T_P_elem',
#            'T_T_elem',
#            'T_H_elem',
#            'T_CO2_elem',
#            'T_aero_elem',
           
#            'H_P_elem',
#            'H_T_elem',
#            'H_H_elem',
#            'H_CO2_elem',
#            'H_aero_elem',
           
#            'CO2_P_elem',
#            'CO2_T_elem',
#            'CO2_H_elem',
#            'CO2_CO2_elem',
#            'CO2_aero_elem',
           
#            'aero_P_elem',
#            'aero_T_elem',
#            'aero_H_elem',
#            'aero_CO2_elem',
#            'aero_aero_elem'
#           ]

columns = ['P_P_elem', 
           'P_T_elem', 
           'P_H_elem',
           'P_aero_elem',
           
           'T_P_elem',
           'T_T_elem',
           'T_H_elem',
           'T_aero_elem',
           
           'H_P_elem',
           'H_T_elem',
           'H_H_elem',
           'H_aero_elem',

           'aero_P_elem',
           'aero_T_elem',
           'aero_H_elem',
           'aero_aero_elem'
          ]
n_samples = len(data_with_lags)

print("Porcentaje de aparición de cada timelag en cada relación (≥ 50%):")
print("-" * 70)

# Iterar por cada relación
for col in columns:
    all_lags = []
    
    for lag_str in data_with_lags[col]:
        lags = literal_eval(lag_str)
        all_lags.extend(lags)
    
    lag_counter = Counter(all_lags)
    
    # Imprimir solo los lags que aparecen en ≥ 50% de las muestras
    for lag_value, count in lag_counter.items():
        percentage = (count / n_samples) * 100
        if percentage >= 1:
            relation = col.replace('_elem', '').replace('_', ' → ')
            print(f"{relation} | lag = {lag_value:2d} → {percentage:.2f}% ({count}/{n_samples})")
```

```python
n_samples = len(data_with_lags)

results = []

# Iterar por cada relación
for col in columns:
    all_lags = []

    for lag_str in data_with_lags[col]:
        lags = literal_eval(lag_str)
        all_lags.extend(lags)

    lag_counter = Counter(all_lags)

    # Guardar solo los lags que aparecen en ≥ 1% de las muestras
    for lag_value, count in lag_counter.items():
        percentage = (count / n_samples) * 100
        if percentage >= 1:
            relation = col.replace('_elem', '').replace('_', ' → ')
            results.append({
                'Relation': relation,
                'Lag': lag_value,
                'Percentage': round(percentage, 2),
                'Count': count,
                'Total Samples': n_samples
            })

# Convert to DataFrame
df_results = pd.DataFrame(results)

# Save to CSV
#df_results.to_csv(path + f'{site}_lag_summary_morning.csv', index=False)
df_results.to_csv(path + f'{site}_lag_summary_rain.csv', index=False)
```
