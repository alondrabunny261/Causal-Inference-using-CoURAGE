```python
#CSV files
site = "S3"
path = f"{site}_data/{site}_Metw_CoURAGE_data"
csv_files = sorted([f for f in os.listdir(path) if f.endswith(".csv")])

all_temp = []
all_rh = []
all_precip = []
all_cumul_precip = []
all_base_time = []
all_atmos_press = []
all_wind = []
all_time = []


for i in csv_files:
    df = pd.read_csv(f"{path}/{i}")
    df = df[['wxt_precip_rate_mean', 
             'rh_mean', 
             'temp_mean', 
             'atmos_pressure', 
             'base_time',
             'wspd_arith_mean',
             'wxt_cumul_precip',
             'time']]
    
    all_precip.extend(df['wxt_precip_rate_mean'])
    all_cumul_precip.extend(df['wxt_cumul_precip'])
    all_rh.extend(df['rh_mean'])
    all_temp.extend(df['temp_mean'])
    all_base_time.extend(df['base_time'])
    all_time.extend(df['time'])
    all_atmos_press.extend(df['atmos_pressure'])
    all_wind.extend(df['wspd_arith_mean'])
    
time = all_time

# Create final DataFrame
met_df = pd.DataFrame({
    'time': all_time,
    'wxt_precip_rate_mean': all_precip,
    'wxt_cumul_precip': all_cumul_precip,
    'rh_mean': all_rh,
    'temp_mean': all_temp,
    'base_time': all_base_time,
    'atmos_pressure': all_atmos_press,
    'wspd_arith_mean':all_wind
})

# # Save to CSV
#met_df.to_csv(f"{site}_minute_meteorological_with_aerosols_data_wxt.csv", index=False)

#############################################################################################

#From metpy package: Create a new column for data that contains dew point and specific humidity

# Calculate dew point using your formula : T - ((100- RH)/5)
met_df['dew_point'] = met_df['temp_mean'] - ((100 - met_df['rh_mean']) / 5)

# Convert pressure from kPa to hPa
pressure = (met_df['atmos_pressure'] * 10).values * units.hPa

# Convert dew point to proper units
dew_point = met_df['dew_point'].values * units.degC

# Compute specific humidity
sh = shfd(pressure, dew_point).to('g/kg')

# Add to dataframe
met_df['specific_humidity'] = sh.magnitude  # only store the values, not units

###############################################################################################

# # Save updated CSV
#df.to_csv(f"{site}_minute_meteorological_with_aerosols_data_wxt.csv", index=False)
```

```python
# CSV File
met_files = sorted([f for f in os.listdir("S3_data/S3_Metw_CoURAGE_data") if f.endswith(".csv")])
co2_files = sorted([f for f in os.listdir("S3_data/S3_Ecorsf_CoURAGE_data") if f.endswith(".csv")])
aerosol_files = sorted([f for f in os.listdir("S3_data/S3_pops_CoURAGE_data") if f.endswith(".csv")])
```

```python
############################ MET DATA ################################

met_day= []
met_time = []

met_precipitation = []
met_cumul_precipitation = []
met_atmos_pressure = []
met_winds = []

met_relative_humid = []
met_relative_humid_std = []

met_temperature = []
met_temperature_std = []


a=0
for file in met_files:
    met_data = pd.read_csv(f"S3_data/S3_Metw_CoURAGE_data/{file}")
    
    ####################### Time Information #############################
    
    day = met_data['base_time']
    time = met_data['time']
    
    ####################### Meteorological Varibles #######################
    
    precipitation  = met_data['wxt_precip_rate_mean']
    cumulative_precipitation = met_data['wxt_cumul_precip']
    atmos_pressure = met_data['atmos_pressure']
    wind_speed = met_data['wspd_arith_mean']
    
    
    relative_humid = met_data['rh_mean']
    relative_humid_std = met_data['rh_std']
    
    temperature    = met_data['temp_mean'] 
    temperature_std = met_data['temp_std'] 
    
    ########################################################################
    
    met_day.extend(day)
    met_time.extend(time)
    
    met_precipitation.extend(precipitation)
    met_relative_humid.extend(relative_humid)
    met_relative_humid_std.extend(relative_humid_std)
    met_temperature.extend(temperature)
    met_temperature_std.extend(temperature_std)
    met_atmos_pressure.extend(atmos_pressure)
    met_winds.extend(wind_speed)

met_df = pd.DataFrame({
    'time': [datetime.strptime(t, "%Y-%m-%d %H:%M:%S") for t in met_time],
    'precipitation': met_precipitation,
    'relative humidity': met_relative_humid,
    'relative humidity std': met_relative_humid_std,
    'temperature': met_temperature,
    'temperature std': met_temperature_std,
    'atmospheric pressure': met_atmos_pressure,
    'wind speed': met_winds
})

# Round to 30 min intervals
met_df['rounded_time'] = met_df['time'].dt.floor('30T')

met_df_avg = met_df.groupby('rounded_time').mean().reset_index()

# Update variables and average every 30 mind
update_days   = met_df_avg['rounded_time'].tolist()
update_precip = met_df_avg['precipitation'].tolist()
update_rh     = met_df_avg['relative humidity'].tolist()
update_temp   = met_df_avg['temperature'].tolist()
update_press  = met_df_avg['atmospheric pressure'].tolist()
update_wind   = met_df_avg['wind speed']
 
update_rh_std   = met_df_avg['relative humidity std'].tolist()
update_temp_std = met_df_avg['temperature std'].tolist()

```

```python
############################ CO2 DATA ################################

co2_day  = []
co2_time = []

co2_co2_flux = []
co2_h2o_flux = []
co2_latent_flux = []


co2_molar_density = []
co2_molar_fraction = []
co2_co2_timelag = []

co2_h2o_molar_density = []
co2_h2o_mole_fraction = []
co2_h2o_timelag = []

co2_evapotranspiration_flux = []
co2_water_vapor_density = []

co2_specific_humidity = []
co2_relative_humidity = []

a=0

for file in co2_files:
    
    co2_data = pd.read_csv(f"S3_data/S3_Ecorsf_CoURAGE_data/{file}")
    
     ####################### Time Information #############################
    
    day = co2_data['base_time']
    time = co2_data['time']
    
    ####################### Varibles #######################
    
    co2_flux                = co2_data['co2_flux']
    co2_mdens               = co2_data['co2_molar_density']
    co2_frac                = co2_data['co2_molar_fraction']
    co2_timelag             = co2_data['co2_timelag']
    
    h2o_flux                = co2_data['h2o_flux']
    h2o_mdens               = co2_data['h2o_molar_density']
    h2o_frac                = co2_data['h2o_mole_fraction']
    h2o_timelag             = co2_data['h2o_timelag']
    
    latent_flux             = co2_data['latent_flux']
    specific_humidity       = co2_data['specific_humidity']
    relative_humidity       = co2_data['relative_humidity']
    water_vapor_density     = co2_data['water_vapor_density']
    evapotranspiration_flux = co2_data['evapotranspiration_flux']

    ########################################################################

    co2_day.extend(day)
    co2_time.extend(time)
    
    co2_co2_flux.extend(co2_flux)
    co2_molar_density.extend(co2_mdens)
    co2_molar_fraction.extend(co2_frac)
    co2_co2_timelag.extend(co2_timelag)
    
    co2_h2o_flux.extend(h2o_flux)
    co2_h2o_molar_density.extend(h2o_mdens)
    co2_h2o_mole_fraction.extend(h2o_frac)
    co2_h2o_timelag.extend(h2o_timelag)
    
    co2_evapotranspiration_flux.extend(evapotranspiration_flux)
    co2_water_vapor_density.extend(water_vapor_density)

    co2_specific_humidity.extend(specific_humidity)
    co2_relative_humidity.extend(relative_humidity)
    
    co2_latent_flux.extend(latent_flux )

    a = a+1

co2_time_dt = [datetime.strptime(t, "%Y-%m-%d %H:%M:%S") if isinstance(t, str) else t for t in co2_time]

```

```python
#Verify consistency in dates

print("Lengths:", len(update_days), len(co2_time_dt))
for i, (u, s) in enumerate(zip(update_days, co2_time_dt)):
    if u != s:
        print(f"Difference at index {i}:")
        print(f"update_days[{i}] = {u} ({type(u)})")
        print(f"co2_time_dt[{i}] = {s} ({type(s)})")
```

```python
############################ Aerosol DATA ################################

pops_day  = []
pops_time = []

pops_total_N_conc           = []
pops_total_count            = []
pops_total_count_std        = []
pops_total_volume_PSL       = []
pops_total_surface_area_PSL = []


a=0

for file in aerosol_files:
    
    aerosol_data = pd.read_csv(f"S3_data/S3_pops_CoURAGE_data/{file}")
    
     ####################### Time Information #############################
    
    day = aerosol_data['base_time']
    time = aerosol_data['time']
    
    ####################### Varibles #######################
    
    N_count         = aerosol_data["total_N_conc"]
    total_count     = aerosol_data["total_count"]
    total_count_std = aerosol_data["total_count_std"]
    total_SA_conc   = aerosol_data["total_SA_conc"]
    total_V_conc    = aerosol_data["total_V_conc"]
    
    ########################################################################
    
    pops_day.extend(day)
    pops_time.extend(time)

    pops_total_N_conc.extend(N_count)
    pops_total_count.extend(total_count)
    pops_total_count_std.extend(total_count_std)
    pops_total_volume_PSL.extend(total_SA_conc )
    pops_total_surface_area_PSL.extend(total_V_conc)



    a = a+1

pops_time_dt = [datetime.strptime(t, "%Y-%m-%d %H:%M:%S") if isinstance(t, str) else t for t in pops_time]

pops_df = pd.DataFrame({
    'time': [datetime.strptime(t, "%Y-%m-%d %H:%M:%S") for t in pops_time],
    'total number conc': pops_total_N_conc,
    'total count': pops_total_count,
    'total count std': pops_total_count_std,
    'total volume PSL': pops_total_volume_PSL,
    'total surface area': pops_total_surface_area_PSL,

})

# Round to 30 min intervals
pops_df['rounded_time'] = pops_df['time'].dt.floor('30T')

pops_df_avg = pops_df.groupby('rounded_time').mean().reset_index()

# Update variables and average every 30 mind
update_days   = pops_df_avg['rounded_time'].tolist()
update_total_number_conc = pops_df_avg['total number conc'].tolist()
update_total_count = pops_df_avg['total count'].tolist()
update_total_count_std = pops_df_avg['total count std'].tolist()
update_total_volume_PSL = pops_df_avg["total volume PSL"].tolist()
update_total_surface_area = pops_df_avg['total surface area'].tolist()
```
