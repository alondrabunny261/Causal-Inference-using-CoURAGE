```python
site = "M1"

path = f"{site}_data/"

# Rain Events
rainy_times = pd.read_csv(path + f"{site}_30_min_extended_rain_windows.csv")

events = list(zip(
    pd.to_datetime(rainy_times['extended_start']),
    pd.to_datetime(rainy_times['extended_end'])))

########################################################################################

only_rainy_times = pd.read_csv(path + f"{site}_30_min_continuous_rain_periods.csv")

start_r = rainy_times.extended_start
end_r   = rainy_times.extended_end

only_r_events = list(zip(
    pd.to_datetime(only_rainy_times['start_time']),
    pd.to_datetime(only_rainy_times['end_time'])))
```

```python
a=0

site = "M1"

path = f"{site}_data/"

#################################################################################################

for rain in events:

    var_names = ['Precipitation','Temperature', "Wetness"]
    
    data = pd.read_csv(path + f"{site}_30_min_meteorological_data.csv")
    data['time'] = pd.to_datetime(data['time'])
    data.set_index('time', inplace=True)

    # Rango del evento (sin resample)
    start_time, end_time = rain
    data_event = data[start_time:end_time]

    ##############################################################################################

    # Selección de variables
    numeric = data_event[["precipitation", "temperature", "wetness"]]

    # Verificación de valores negativos
    negative_precip_mask = numeric["precipitation"] < 0
    print(f"Negative precipitation values in event {a}: {negative_precip_mask.sum()}")

    # Convertir a array
    data_array = numeric.to_numpy()

    # Reemplazo de valores negativos y NaNs
    data_array[negative_precip_mask.to_numpy(), 0] = 999999 
    data_array[np.isnan(data_array)] = 999999

    T, N = data_array.shape
    
    # Crear Tigramite DataFrame
    dataframe = pp.DataFrame(data_array, var_names=var_names, missing_flag=999999)

    ######################################################################################################

    ############################### Plotting Dataframe ####################################################

    tp.plot_timeseries(dataframe)
    plt.show()
    
    ########################################################################################################
    
    parcorr = ParCorr(significance='analytic')

    pcmci = PCMCI(
    dataframe=dataframe, 
    cond_ind_test=parcorr,
    verbosity=1)
    
    correlations = pcmci.get_lagged_dependencies(tau_max=2, val_only=True)['val_matrix']
    
    matrix_lags = None #np.argmax(np.abs(correlations), axis=2)
    
    #tp.plot_scatterplots(dataframe=dataframe, add_scatterplot_args={'matrix_lags':matrix_lags}); plt.show()
    
    tp.plot_densityplots(dataframe=dataframe, add_densityplot_args={'matrix_lags':matrix_lags})
    plt.show()
    
    correlations = pcmci.get_lagged_dependencies(tau_max=2, val_only=True)['val_matrix']

    lag_func_matrix = tp.plot_lagfuncs(val_matrix=correlations, setup_args={'var_names':var_names, 
                                                                            'x_base':5, 
                                                                            'y_base':.5})
    plt.show()
    
    pcmci.verbosity = 1
    results = pcmci.run_pcmci(tau_max=2, pc_alpha=None, alpha_level=0.05)
    
    print("p-values")
    print (results['p_matrix'].round(3))
    print("MCI partial correlations")
    print (results['val_matrix'].round(2))
    
    tau_max=2

    q_matrix = pcmci.get_corrected_pvalues(p_matrix=results['p_matrix'], tau_max=tau_max, fdr_method='fdr_bh')

    pcmci.print_significant_links(
            p_matrix = q_matrix,
            val_matrix = results['val_matrix'],
            alpha_level = 0.05)
    graph = pcmci.get_graph_from_pmatrix(p_matrix=q_matrix, 
                                         alpha_level=0.05, 
                                         tau_min=0, 
                                         tau_max=tau_max, 
                                         link_assumptions=None)
    results['graph'] = graph
    
    ########################################## Plotting ##############################################
    
    tp.plot_graph(val_matrix=results['val_matrix'],
                  graph=results['graph'],
                  var_names=var_names,
                  link_colorbar_label='cross-MCI',
                  node_colorbar_label='auto-MCI',
                  show_autodependency_lags=False)
    plt.show()
    
    ###################################################################################################
    
    tau_max = 2
    pc_alpha = 0.05
    link_assumptions = None
    results = pcmci.run_pcmciplus(tau_max=tau_max, 
                                  pc_alpha=pc_alpha,
                                  link_assumptions=link_assumptions,
                                 )
    tp.plot_graph(
        val_matrix=results['val_matrix'],
        graph=results['graph'],
        ); plt.show()
    
    # Inicializa todos los links como bloqueados (None = no existe)
    link_assumptions = {
        j: {} for j in range(N)
    }

    # Solo permite: Precipitation (0) --> Wetness (2) at lag = 1
    link_assumptions[2][(0, -1)] = '-->'
    results = pcmci.run_pcmciplus(tau_max=tau_max, 
                                  pc_alpha=pc_alpha,
                                  link_assumptions=link_assumptions,
                                 )
    print(results['graph'][:,:,0])
#     tp.plot_graph(
#         val_matrix=results['val_matrix'],
#         graph=results['graph'],
#         ); plt.show()
    
    tp.plot_graph(
    val_matrix=results['val_matrix'],
    graph=results['graph'],
    var_names=var_names,
    link_colorbar_label='cross-MCI',
    node_colorbar_label='auto-MCI',
    show_autodependency_lags=True
    )
```
