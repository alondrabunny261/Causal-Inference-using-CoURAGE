```python
site = "S3"
path = f"{site}_data/30_min_CO2_aerosols/"

# Morning
if site == "M1":
    sample_size = np.arange(0, 176, 1)
    
elif site == "S2":
    sample_size = np.arange(0, 162, 1)

elif site == "S3":
    sample_size = np.arange(0, 180, 1)
    
```

```python
data_morning = pd.read_csv(path + f"WPC_{site}_evening_table.csv")


pair_metadata = {
    'CO2_T': {'label': 'CO2→Temperature',   'color': 'cyan'},
    'Aerosols_T': {'label': 'Aerosols→Temperature',     'color': 'plum'},
    'Specific_Humidity_T': {'label': 'Specific Humidity→Temperature',     'color': 'orchid'},
    'Specific_Humidity_Aerosols': {'label': 'Specific Humidity→Aerosols',     'color': 'skyblue'} }
```

```python

# Convierte la columna de strings a números reales
def extract_number(value):
    try:
        return float(ast.literal_eval(value)[0])  # extrae el primer valor de la lista
    except:
        return None  # o np.nan si prefieres

# Aplica la conversión a cada columna relevante
for col in data_morning.columns:
    data_morning[col] = data_morning[col].apply(extract_number)
```

```python
data_morning
```

```python

# Extraer los datos de interés
effect = data_morning['CO2 on Temperature']
interval_low = data_morning["CO2 on Temperature interval 1"]
interval_high = data_morning["CO2 on Temperature interval 2"]

# Configuración desde pair_metadata
color = pair_metadata['CO2_T']['color']
label = pair_metadata['CO2_T']['label']

# Crear el histograma del efecto
plt.figure(figsize=(8, 5))
plt.hist(effect, bins=20, color=color, edgecolor='black', alpha=0.7)
plt.axvline(effect.mean(), color='blue', linestyle='dashed', linewidth=1.5, label='Mean')
plt.text(-15,10, f"Mean : {np.mean(effect)}")

# Título y etiquetas
plt.title(f"{site} Facility Evening: Effect {label}")
plt.xlabel("Magnitud ")
plt.ylabel("Frecuency")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
